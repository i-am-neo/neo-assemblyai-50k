# An intro to Neo

For remote teams, Neo is an AI who functions as an extra teammate. He's designed to become the knowledge repository for the team, to inform, notify, and potentially in the future, be trusted enough by the team to update documents.  

Neo is designed to have "eyes" and "ears" - for this AI-first challenge, we show you the kind of information he can glean - this example from a team meeting.


# An example output

Takeaways can be organized to more closely resemble a team agenda which can be sent to Neo with your calendar invite.

More importantly, the key takeways can be organized in a personalized way.  For example, if you work on security-related projects in your team, you can ask Neo to watch for "security" and he will point you to those takeaways first, if they were discussed in your team meeting.

# Neo compiles the information into a "recap"
which he sends to each teammember in the calendar invite.  Each recap will be personalized accordingly (name mentions + any specified watchlists).
